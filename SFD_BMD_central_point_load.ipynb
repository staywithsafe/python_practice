```python
# 단순보 + 중앙 집중하중
P = 20.0   # 집중하중 (kN)
L = 6.0    # 경간 (m)

print(f"P = {P} kN, L = {L} m")
```

```python
# 좌표 분할
n = 200
x_list = []
V_list = []
M_list = []

for i in range(n + 1):
    x = L * i / n
    x_list.append(x)

    # 전단력 (SFD)
    if x < L / 2:
        V = P / 2
    elif x > L / 2:
        V = -P / 2
    else:
        V = 0.0
    V_list.append(V)

    # 휨모멘트 (BMD)
    if x <= L / 2:
        M = (P / 2) * x
    else:
        M = (P / 2) * (L - x)
    M_list.append(M)
```

```python
import matplotlib.pyplot as plt

# SFD
plt.figure()
plt.plot(x_list, V_list)
plt.axhline(0)
plt.title("SFD - Central Point Load")
plt.xlabel("x (m)")
plt.ylabel("V (kN)")
plt.grid(True)

# BMD
plt.figure()
plt.plot(x_list, M_list)
plt.axhline(0)
plt.title("BMD - Central Point Load")
plt.xlabel("x (m)")
plt.ylabel("M (kN·m)")
plt.grid(True)

plt.show()
```
